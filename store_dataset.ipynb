# Shipment case study:	


```python
import pandas as pd
import numpy as np
import math
import numpy as kp
```

```python
store = pd.read_csv("E:\Store.csv",sep=',',header=0,encoding="latin")
store
```

# How many unique cities are the orders being delivered to

```python
store.City.nunique()
```

# 	What is the total quantity sold in the East Region?

```python
store[store.Region=='East'].count().Quantity
```

# Find the sum of the quantity sold in the East Region

```python
store[store.Region == 'East'].agg({'Quantity':np.sum})
```

# 	In the south region sort the sales in decreasing order

```python
a = store[store.Region == 'South']
sorted(a.Sales,reverse=True)
```

# 	Find the mean of quantity for every region

```python
store.groupby('Region').agg({'Quantity':np.mean})
```

# 	Find the mean of sales for every category

```python
store.groupby('Category').agg({'Sales':np.mean})
```

# 	Find the max, min, sum of sales and profit for every category

```python
store.groupby('Category')[['Sales','Profit']].agg(['max','min','sum'])
```

# 	Find sum of sales and max profit for every segment

```python
store.groupby('Segment').agg({'Sales':np.sum,'Profit':np.max})
```

# 	For every segment find the mean of the discount

```python
store.groupby('Segment').agg({'Discount':np.mean})
```

# 	For every segment find the most profitable customers

```python
a=[store.groupby("Segment").max().Profit]
b = [store.Profit]
store[["Customer Name","Segment"]][kp.in1d(b,a)]
```

# 	What are the top 5 categories that give maximum profit?

```python
store.groupby('Sub-Category').max()['Profit'].sort_values(ascending = False).head(5)
```

# 	What is the Total Sales, Quantity, Discount, Profit across Total US.

```python
store.groupby('Country').agg({'Sales':np.sum,'Quantity':np.sum,'Discount':np.sum,'Profit':np.sum})
```

# 	How many times has it taken more than 5 days from placing an order to shipping

```python
store['Order Date']=pd.to_datetime(store['Order Date'])
store['Ship Date']=pd.to_datetime(store['Ship Date'])
count=(store['Ship Date']-store['Order Date'])>"5 days"
sum(count==True)
```

# 	Find the total number of orders in every category which has been shipped with a duration > 5 days

```python
store[(store['Ship Date']-store['Order Date'])>'5 days'].groupby('Category')['Customer ID'].count()
```

# 	What’s the percentage of items which has been shipped within 5 days

```python
a=store[(store['Ship Date']-store['Order Date'])<='5 days']['Customer ID'].count()
b=len(store['Ship Date']-store['Order Date'])
round(a/b*100,2)
```

# 	What’s the percentage of items which has been shipped after 5 days

```python
c=store[(store['Ship Date']-store['Order Date'])>'5 days']['Customer ID'].count()
round(c/b*100,2)
```
